```python
import pandas as pd
import numpy as np
```

```python
# valin harjutamiseks Harry Kane'i eelmise hooaja
path = "./Fantasy-Premier-League-master/data/2019-20/players/Harry_Kane_338/gw.csv"
```

```python
# loen faili sisse
stats = pd.read_csv(path)
# näitan kõiki tulpasid
pd.set_option("display.max_columns", None)
```

```python
stats.head(10)
```

```python
# viskan välja enda hinnangul ebavajalikud tulbad:
# element - mängija ID
# fixture - mängu ID
# kickoff_time - mängu algus
# round - raundi number
# selected - mitu inimest mängija valinud on
# transfers_in - mitu inimest juurde võttis eelmise nädalaga võrreldes
# transfers_out - mitu inimest ära andis eelmise nädalaga võrreldes
# transfers_balance - eelmiste vahe

stats_edited = stats.drop(["element", "fixture", "kickoff_time", "round", "selected", "transfers_in", "transfers_out", "transfers_balance"], axis=1)
```

```python
stats_edited.head(10)
```

```python
# hakkan looma järgmise nädala punktide listi
# eraldan mängijate punktid nädalate kaupa
points_next_week = list(stats_edited["total_points"])
# eemaldan esimese nädala skoori, sest seda ei ennustata millegi põhjal
del points_next_week[0]
# käimasoleva nädala lahtrisse panen -1, sest seda ei oska öelda
points_next_week.append(-1)
print(points_next_week)
# lisan dataframe'i
stats_edited["points_next_week"] = points_next_week
```

```python
stats_edited.head(10)
```

```python
# eraldan kasutatavad ja ennustatava tunnuse
X = stats_edited.drop("points_next_week", axis=1)
y = stats_edited.points_next_week

# võtan esimesed 30 nädalat testhulgaks ja viimased 8 treeninghulgaks
X_train = X.iloc[:30]
y_train = y.iloc[:30]
X_test = X.iloc[30:].reset_index(drop=True)
y_test = y.iloc[30:].reset_index(drop=True)
```

```python
# proovin esialgu tavalist random forest regressorit
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor().fit(X_train, y_train)
predictions = model.predict(X_test)

for i in range(len(predictions)):
    print("Ennustatud:", round(predictions[i],2),"| Tegelik:", y_test[i])
```
